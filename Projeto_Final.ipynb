```python
import pandas as pd
import glob

# Lista todos os arquivos CSV no diretório especificado
arquivos = glob.glob('/content/content/*.csv')

# Cria uma lista para armazenar os DataFrames
lista_df = []

for arquivo in arquivos:
    try:
        df = pd.read_csv(arquivo, sep=';', encoding='latin1', low_memory=False)
        lista_df.append(df)
        print(f"Arquivo lido: {arquivo}")
    except Exception as e:
        print(f"Erro ao ler {arquivo}: {e}")

# Verifica se há dados para concatenar
if lista_df:
    # Concatena os DataFrames
    df_consolidado = pd.concat(lista_df, ignore_index=True)

    # Salva como UTF-8 com BOM (compatível com Excel)
    df_consolidado.to_csv('execucao_despesa_consolidado.csv', index=False, encoding='utf-8-sig')
    print("Arquivo consolidado salvo com sucesso.")
else:
    print("Nenhum arquivo foi carregado corretamente.")
```
